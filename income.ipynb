```python
import pandas as pd
import matplotlib as plt
```

```python
df = pd.read_csv('income.csv')
print(df.head())

print(len(df.index))
```

```python
df = df.drop('Age', axis=1)
print(df.head()) 
print(df.columns)
```

```python
df = df.map(lambda x: x.strip() if isinstance(x, str) else x) # remove white spaces
df = df.dropna() # remove blanks
```

```python
df['Income (Annual)'] = df['Income (Annual)'].str.replace('$', '').str.replace(',', '').astype(int)
df['Classification'] = df['Classification'].apply(lambda x: '1' if x == 'Saver' else '2' if x == 'Budget-Conscious' else '3' if x == 'Spender' else None).astype(int)
print(df.head())
```

```python
df.plot.scatter(x='Spending Score (1-100)', y='Income (Annual)', c='Classification', colormap='viridis')
```
