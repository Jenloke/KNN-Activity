```python
import pandas as pd
import matplotlib as plt
```

Reading of CSV

```python
df = pd.read_csv('income.csv')
print(df.head())

print(len(df.index))
```

```python
df = df.drop('Age', axis=1)
print(df.head()) 
```

```python
print(df.columns)
```

```python
df = df.map(lambda x: x.strip() if isinstance(x, str) else x) # remove white spaces
df = df.dropna() # remove blanks

print(df.head())
```

```python
df['Income (Annual)'] = df['Income (Annual)'].str.replace('$', '').str.replace(',', '').astype(int)
print(df.head())
```

```python
df.plot.scatter(x='Spending Score (1-100)', y='Income (Annual)')
```
